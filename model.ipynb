```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("features_3_sec.csv")
```

```python
df.head()
```

```python
df.drop(columns=['mfcc1_mean','mfcc2_mean','mfcc3_mean','mfcc4_mean','mfcc5_mean','mfcc6_mean','mfcc7_mean','mfcc8_mean','mfcc9_mean','mfcc10_mean','mfcc11_mean','mfcc12_mean','mfcc13_mean','mfcc14_mean','mfcc15_mean','mfcc16_mean','mfcc17_mean','mfcc18_mean','mfcc19_mean','mfcc20_mean','filename','length'], inplace=True)
```

```python
df.head()
```

```python
print(df.columns)
len(df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
encoder = LabelEncoder()
scaler = StandardScaler()
data = df.iloc[:, :-1] 
data
```

```python
labels = df.iloc[:, -1] # get labels column
labels.to_frame()
```

```python
x = np.array(data, dtype = float)
x = scaler.fit_transform(data)
y = encoder.fit_transform(labels)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier

KN=KNeighborsClassifier()

KN.fit(x_train,y_train)
z=KN.predict(x_test)
#for i in z:
print(encoder.inverse_transform(z))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,KN.predict(x_test))
```

```python
import pickle
pick1 = {
    'scaler':scaler,
    'knn':KN,
    'encoder':encoder
}
pickle.dump( pick1, open( 'mlmodel' + ".p", "wb" ) )
```
